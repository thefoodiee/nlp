importing libraries and downloading important resources


```python
# importing libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer

# download required nltk resources
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

Defining the paragraph.

```python
paragraph = (
    """
Arch Linux is a lightweight and flexible Linux distribution that emphasizes simplicity and customization. Unlike many other distributions, Arch follows a rolling release model, meaning it’s continuously updated rather than requiring major version upgrades. This gives users access to the latest software, but also demands a higher level of involvement in maintaining the system. The Arch Wiki is widely regarded as one of the most comprehensive resources available, providing detailed documentation for users at all experience levels. Arch's minimalist approach allows users to build their systems from the ground up, choosing only the components they need. While this offers tremendous flexibility, it also means that beginners may face a steep learning curve compared to more user-friendly distributions. Despite these challenges, Arch has gained a dedicated following, as it provides unmatched control over the system and allows for an optimized, tailored experience.
"""
)
print(paragraph)
```

Tokenization : This splits the paragraph into individual words and punctuation marks


```python
tokens = word_tokenize(paragraph)
print("Token")
print(tokens)
```

Stopword Removal : This removes common english words like "is","the","and",etc. Keeping only meaningful words.

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
print("after stopword removal")
print(filtered_tokens)
```

Stemming : Reduce a word to its root forms often not actual words

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("after stemming")
print(stemmed_words)
```

Lemmatization : lemmatization returns dictionary base forms.

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("After lemmatization")
print(lemmatized_words)
```
